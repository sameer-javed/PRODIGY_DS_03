```python
# Creating X and Y variables 
x = df.drop(columns={'y'})
y = df['y']
```

```python
x.columns
```

# Label Encoding the Categorical data 

```python
le = LabelEncoder()
x['age'] = le.fit_transform(x['age'])
x['job'] = le.fit_transform(x['job'])
x['marital'] = le.fit_transform(x['marital'])
x['education'] = le.fit_transform(x['education'])
x['default'] = le.fit_transform(x['balance'])
x['housing'] = le.fit_transform(x['housing'])
x['loan'] = le.fit_transform(x['loan'])
x['contact'] = le.fit_transform(x['contact'])
x['day'] = le.fit_transform(x['day'])
x['month'] = le.fit_transform(x['month'])
x['duration'] = le.fit_transform(x['duration'])
x['campaign'] = le.fit_transform(x['campaign'])
x['pdays'] = le.fit_transform(x['pdays'])
x['previous'] = le.fit_transform(x['previous'])
x['poutcome'] = le.fit_transform(x['poutcome'])
```

```python
x
```

```python
y = le.fit_transform(y)
```

```python
y
```

# Splitting the data to train test split

```python
X_train , X_test , Y_train , Y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

# Creating a model Using Decision Tree Classifier

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```
